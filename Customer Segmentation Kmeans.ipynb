```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
customer_data = pd.read_csv('/content/Mall_Customers.csv')
```

```python
customer_data.head()
```

```python
customer_data.shape
```

```python
customer_data.info()
```

```python
#separating annual income and spending score columns from the rest
x = customer_data.iloc[:,[3,4]].values
```

```python
print(x)
```

Choosing correct no. of clusters

WCSS--> within cluster sum of squares

```python
#finding wcss value for diff no. of clusters
wcss=[]

for i in range(1,11) :
  kmeans=KMeans(n_clusters=i, init= 'k-means++', random_state= 23)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
#plotting elbow graph

sns.set()
plt.plot(range(1,11),wcss)
plt.title("Elbow Point graph")
plt.xlabel("No. of clusters")
plt.ylabel("WCSS value")
plt.show()
```

Optimum no. of clusters = 5

```python
#training model with 5 clusters

kmeans= KMeans(n_clusters=5, init= 'k-means++', random_state=1)

#storing labels of all data points based on their cluster
y= kmeans.fit_predict(x)
print(y)
```

```python
#plotting all the clusters and centroids

plt.figure(figsize= (8,8))
plt.scatter(x[y==0,0], x[y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(x[y==1,0], x[y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(x[y==2,0], x[y==2,1], s=50, c='blue', label='Cluster 3')
plt.scatter(x[y==3,0], x[y==3,1], s=50, c='yellow', label='Cluster 4')
plt.scatter(x[y==4,0], x[y==4,1], s=50, c='violet', label='Cluster 5')

#plotting centriods
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=70, c='black', label='Centroid')

plt.title('Customer Groups')
plt.xlabel('Annual income')
plt.ylabel('Spending score')
plt.show()
```
